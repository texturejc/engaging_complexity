# The fold catastrophe

The fold catastrophe has the form: $$\large{V = x^3 + ax}$$

The catastrophe surface is the set of points for which the rate of change of $V$ with respect to $x = 0$, giving: $$\large{\frac{dV}{dx} = 3x^2 + a = 0}$$ as the catastrophe surface.(Note that $\frac{dV}{dx}$ is the rate of change.)

```python
r = [i for i in np.arange (-10, 10, 0.001)] #numerical range of the data

def fold_p(x): # fold catastrophe for a > 0
    y = x**3 + 40*x
    return y

def fold_n(x): # fold catastrophe for a < 0
     y = x**3 - 40*x
     return y

def fold_0(x): # fold catastrophe for a = 0
     y = x**3
     return y

def s(x): # Fold catastrophe surface
    y = -3*x**2
    return y

```

# Fold catastrophe for a < 0

Note the presence of two critical points: an unstable upper point and a stable lower point

```python
plt.plot(r, [fold_n(i) for i in r])
plt.xlabel('x')
plt.ylabel('V')
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.title('Fold catastrophe for a < 0')
```

# Fold catastrophe for a = 0

Note the how the two critical points merge at $a = 0$

```python
plt.plot(r, [fold_0(i) for i in r])
plt.xlabel('x')
plt.ylabel('V')
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.title('Fold catastrophe for a = 0')
```

# Fold catastrophe for a > 0

Note how that there is now no critical point; the system has transitioned into a new behaviour and no return is possible (there are no points where the rate of change is 0).

```python
plt.plot(r, [fold_p(i) for i in r])
plt.xlabel('x')
plt.ylabel('V')
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.title('Fold catastrophe for a > 0')
```

# Catastrophe surface of fold catastrophe

The upper and lower branches correspond to the unstable and stable critical points. They meet at $a = 0$

```python
plt.plot([s(i) for i in r], r)
plt.xlabel('a')
plt.ylabel('x')
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')
plt.title('Fold catastrophe surface as a  0')
```

# What do the fold variables mean? Some examples

## Ecological collapse

*x*: The state of an ecological system; ranges from stable (self-regulating, persistent) to unstable (collapse of the system into thermodynamic equilibrium)

*a*: An environmental stressor, in the form of an unintegrated external force

*Causal model*: An ecological system can exist in a stable, self-regulating state or not; no third option exists. This can be represented as a competition bewtween two equilibria: a stable, self-regulating equilibrium and and unstable destructive one. For mild stressors, the system can absorb the shock of the external equilibrium without major effects and thus both can co-exist. However, once a critical threshold is crossed, the system transitions completely into a new state, where a new rgime dominates.

*Example*: Coral reefs are vulnerable to changes in temperature. For small perturbations, these can be tolerated; [for larger perturbations, coral bleaching occurs](https://oceanservice.noaa.gov/facts/coral_bleach.html).

# The cusp catastrophe

The cusp catastrophe has the form: $$\large{V = -x^4 +ax^2 + bx}$$

The catastrophe surface is the set of points for which the rate of change of $V$ with respect to $x = 0$, giving: $$\large{\frac{dV}{dx} = -4x^3 + 2ax + b = 0}$$ as the catastrophe surface.

# What do the cusp variables mean? Two examples

## Black lives matter protests

*x*: the behavioural state of society; ranges from placid to agitated

*a*: the level of poltical polarisation of society; ranges from low levels of intra-social conflict to to high levels of intrasocial conflict.

*b*: is the salience of racist incidents; ranges from low salience to high salience

*Causal model*: For low levels of political polarisation, increased salience of racist incidents brings about a continuous change from placidity through neutrality to agitation. However, as political polarisation levels increase, the neutral option vanishes, and the system has only two states: placidity and agitation. In this scenario, increasing the level of salience causes the system to suddenly jump from placidity to agitation. Once in the agitated state, the system can either become less polarised and go back to its intial state, or oscillate between high and low agitation in respose to increases and decreases in the salience of racism.

*Historical narrative*: Previously, the dominance of white Eurocentric culture in the USA damped the effects of cultural and social movements, thereby lowering widespread agitation. However, the emergence of large, self-aware non-white demographic groups, coupled with growing economic precarity, increased intra-social antagonism. At the same time, the emergence of widespread social media sharing made incidents of discrimination and violence perceptually salient. The murder of George Floyd caused the system to jump from placid to agitated by making a particularly egregious example of racism visible when social tension was at its highest due to the pandemic.


## Outbreak of World War I

*x*: the behavioural state of society; ranges from peaceful to aggressive

*a*: the level of currently available opportunity that can return from aggressive action; ranges from low levels of return to to high levels of return

*b*: is the weakness of resistance against international conflict; it can be thought of as a combination of domestic and international factors (i.e. popular support for war, robustness of international insitutions).

*Causal model*: For low levels of opportunity, greater degrees of weakness in resitance to conflict brings about a continuous change from peacelike behaviour through neutrality to aggressive behaviour. However, as opportunity levels increase, the neutral option vanishes, and the system has only two states: peaceful and aggressive. In this scenario, increasing the weakness of international norms further causes the system to suddenly jump from peaceful to aggressive. Once in the aggressive state, the system can either become more robust against unilaternal action and go back to its intial state, or oscillate between high and low aggression in respose to increases and decreases in the availability of opportunity.

*Historical narrative*: In the run-up to 1914, European powers faced few international sanctions and had febrile populations who were in favour of expansionism. By 1914, the emergence of the two major European alliances created a first-mover advantage that meant opportunity became more salient. The result was the supervening of war in July-August, where a (relatively) small perturbation (the assassination of Franz Ferdinand) led to a major international conflict.

## An predator chasing prey

*x*: the behavioural state of the predator; ranges from aversion to pursuit to indifference to desire for pursuit

*a*: the level of hunger, indicating the energy need of the predator; ranges from low levels of hunger to to high levels of hunger

*b*: is the ease of chasing and catching the prey; it is essentially a probability function

*Causal model*: For low levels of hunger, greater ease of catching the prey brings about a continuous change from aversion behaviour through indifference to active behaviour. However, as hunger levels increase, the neutral option vanishes, and the predator has only two states: static and active. In this scenario, increasing the ease of capture further causes the predator to suddenly jump from static to active. Once in the active state, the predator can either become satiated and go back to its intial state, or lose energy and return to the aversive state.

*Examples*: There is good zoologica evidence that many predators actively balance the energy expenditure in catching prey against the energy return on success. This is presumably done by balancing internal variables (hunger, fatigue) against the known difficulty of catching certain types of prey and the calorie return on consuming it.
